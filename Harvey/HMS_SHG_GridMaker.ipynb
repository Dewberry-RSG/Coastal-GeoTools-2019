```python
import numpy as np
from matplotlib import pyplot as plt
% matplotlib inline
```

```python
dss_grid = r'T:\CCSI\TECH\2018_HoustonCity\GIS\DATA\Harvey_Rainfall_Reconstruction\ascii\Harvey_reconstructed_totalrain.asc' 
grid = np.loadtxt(dss_grid , skiprows=6)
print(grid.shape)
plt.imshow(grid)
```

```python
# From DSS grid info
llx = -20
lly = 692

# Make SHG Grid coord map
x_coords, y_coords = np.meshgrid(np.arange(llx, llx+grid.shape[0]),np.arange(lly, lly+grid.shape[1]+1), indexing='ij')
```

```python
shg_y_coords = y_coords
shg_y_coords
```

```python
shg_x_coords = np.flip(x_coords, 0)
shg_x_coords
```

```python
shg_grid_x = r'T:\CCSI\TECH\2018_HoustonCity\GIS\DATA\Harvey_Rainfall_Reconstruction\hms_development\shg_grid_x.asc' 

with open(dss_grid, 'r') as fin:
    with open(shg_grid_x, 'w') as fout:
        for i in range(6):
            line = fin.readline()
            print(line)
            fout.write(line)
        for row in shg_x_coords:
            row_data = np.array_str(row)
            row_data = row_data.replace('[','').replace(']','')
            fout.write(row_data+ '\n')
```

```python
shg_grid_y = r'T:\CCSI\TECH\2018_HoustonCity\GIS\DATA\Harvey_Rainfall_Reconstruction\hms_development\shg_grid_y.asc' 

with open(dss_grid, 'r') as fin:
    with open(shg_grid_y, 'w') as fout:
        for i in range(6):
            line = fin.readline()
            print(line)
            fout.write(line)
        for row in shg_y_coords:
            row_data = np.array_str(row)
            row_data = row_data.replace('[','').replace(']','')
            fout.write(row_data+ '\n')
```
